```python
import pandas as pd
import seaborn as sns
sns.set(color_codes=True)
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Sleep_health_and_lifestyle_dataset.csv")
```

```python
df.head()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='Gender', data=df)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Number of Individuals')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(y='Occupation', data=df, order=df['Occupation'].value_counts().index)
plt.title('Occupation Distribution')
plt.xlabel('Number of Individuals')
plt.ylabel('Occupation')
plt.show()
```

```python

plt.figure(figsize=(10, 5))
sns.boxplot(x='Age', y='Sleep Duration', data=df)
plt.title('Sleep Duration by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Sleep Duration (hours)')
plt.show()
```

```python
sleep_disorder_counts = df.groupby(['Gender', 'Sleep Disorder']).size().unstack()
sleep_disorder_counts.plot(kind='bar', stacked=True)
plt.title('Sleep Disorder Status by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Sleep Disorder')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x='Sleep Duration', y='Quality of Sleep', data=df)
plt.title('Sleep Duration vs. Quality of Sleep')
plt.xlabel('Sleep Duration (hours)')
plt.ylabel('Quality of Sleep (1-10)')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(x='Physical Activity Level', y='Heart Rate', data=df)
plt.title('Physical Activity Level vs. Heart Rate')
plt.xlabel('Physical Activity Level')
plt.ylabel('Resting Heart Rate (bpm)')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x='BMI Category', y='Daily Steps', data=df)
plt.title('Distribution of Daily Steps Across BMI Categories')
plt.xlabel('BMI Category')
plt.ylabel('Daily Steps')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
heatmap_data = df.groupby(['Stress Level', 'Quality of Sleep']).size().unstack()
sns.heatmap(heatmap_data, annot=True, cmap='YlGnBu')
plt.title('Heatmap of Stress Levels vs. Sleep Quality')
plt.xlabel('Sleep Quality')
plt.ylabel('Stress Level')
plt.show()
```
